```python
df['city']=cities
```

```python
df.drop(columns=['venue'],inplace=True)
df
```

```python
eligible_cities=df['city'].value_counts()[df['city'].value_counts()>600].index.tolist()
```

```python
df[df['city'].isin(eligible_cities
                  )]
```

```python
df['current_score']=df.groupby('match_id').cumsum()['runs']
df
```

```python
df['over']=df['ball'].apply(lambda x:str(x).split(".")[0])
df['ball_no']=df['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
df['balls_bowled']=(df['over'].astype('int')*6) + df['ball_no'].astype('int')
```

```python
df['balls_left']=120 - df['balls_bowled']
df['balls_left']=df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
df
```

```python
df['player_dismissed']=df['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
df['player_dismissed']=df['player_dismissed'].astype('int')
df['player_dismissed']=df.groupby('match_id').cumsum()['player_dismissed']
df['wickets_left']=10-df['player_dismissed']
df
```

```python
df['current_run_rate']=(df['current_score']*6)/df['balls_bowled']
df
```

```python
groups = df.groupby('match_id')
match_ids = df['match_id'].unique()
last_five = []

for id in match_ids:
    valid_columns = ['runs']  # Add other valid columns as needed

    last_five.extend(groups.get_group(id)[valid_columns].rolling(window=30).sum()['runs'].values.tolist())
```

```python
df['last_five']=last_five
df
```

```python
final_df=df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
final_df
```

```python
final_df=final_df[['batting_team','bowling_team','city','current_score','balls_left','wickets_left','current_run_rate','last_five','runs_x']]
```

```python
final_df = final_df.copy()
final_df.dropna(inplace=True)
```

```python
final_df=final_df.sample(final_df.shape[0])
final_df
```

```python
x=final_df.drop(columns=['runs_x'])
y=final_df['runs_x']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Define your feature columns andfinal_df.drop(columns=['runs_x']) target column
feature_columns=final_df.drop(columns=['runs_x'])
target_columns=final_df['runs_x']



```

```python
x_train, x_test, y_train, y_test = train_test_split(feature_columns, target_columns, test_size=0.2, random_state=1)
```

```python
x_train
```

```python
# Define the preprocessing steps for different types of features
numeric_transformer = StandardScaler()  # Apply standard scaling to numeric features
categorical_transformer = OneHotEncoder(sparse='deprecated', sparse_output=True)  # Apply one-hot encoding to categorical features

# Combine the preprocessing steps with the estimator in a pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, ['current_score','balls_left','wickets_left','current_run_rate','last_five']),
        ('cat', categorical_transformer,  ['batting_team','bowling_team','city'])
    ])

estimator = RandomForestRegressor()  # Replace with your chosen estimator

pipe = Pipeline(steps=[('preprocessor', preprocessor), ('estimator', estimator)])

# Fit the pipeline on the training data
pipe.fit(x_train, y_train)

# Make predictions on the test data
y_pred = pipe.predict(x_test)

# Calculate evaluation metrics
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Print the evaluation metrics
print("R2 Score:", r2)
print("Mean Absolute Error:", mae)
```
